### Create an average of all 'Math', 'Reading', and 'Science' scores <not necessary for each score>

```python
pisa_c['Average_Math_Score']    = (pisa_c['PV1MATH'] + pisa_c['PV2MATH'] + pisa_c['PV3MATH'] + pisa_c['PV4MATH'] + pisa_c['PV5MATH']) / 5

# Create average of Reading scores
pisa_c['Average_Reading_Score'] = (pisa_c['PV1READ'] + pisa_c['PV2READ'] + pisa_c['PV3READ'] + pisa_c['PV4READ'] + pisa_c['PV5READ']) / 5

# Create average of Science scores
pisa_c['Average_Science_Score'] = (pisa_c['PV1SCIE'] + pisa_c['PV2SCIE'] + pisa_c['PV3SCIE'] + pisa_c['PV4SCIE'] + pisa_c['PV5SCIE']) / 5

# Create average score
pisa_c['Average_Total_Score']   = (pisa_c['Average_Math_Score'] + pisa_c['Average_Reading_Score'] + pisa_c['Average_Science_Score']) / 3
```

### Take action for multiple columns for single variables related to father's education

Values for the father's education are distributed among 5 columns accordance to this classification: http://uis.unesco.org/sites/default/files/documents/international-standard-classification-of-education-isced-2011-en.pdf
I will have to expand the first column by creating dummies for each data to have a column, and then melt all the columns into one.

```python
#Make a copy 
f_edu = pisa_c.copy()
```

```python
# Split column up with dummies
f_edu[['<ISCED level 1>',
       '<ISCED level 2>', 
        '<ISCED level 3A>', 
        '<ISCED level 3B, 3C>', 
        'He did not complete <ISCED level 1>']] = pd.get_dummies(f_edu['Father - Highest Schooling'])
```

```python
# Rename the columns for consistency and to remove subcategorization of levels
f_edu.rename({'He did not complete <ISCED level 1>': '<ISCED level 0>',
                   '<ISCED level 3A>': '<ISCED level 3>', 
                   '<ISCED level 3B, 3C>': '<ISCED level 3>',
                   'Father Qualifications - ISCED level 4': '<ISCED level 4>',
                   'Father Qualifications - ISCED level 5A': '<ISCED level 5>',
                   'Father Qualifications - ISCED level 5B': '<ISCED level 5>',
                   'Father Qualifications - ISCED level 6': '<ISCED level 6>'}, axis='columns', inplace=True)
```

```python
# Change datatypes from ints into strings for dummy columns
f_edu['<ISCED level 0>'] = f_edu['<ISCED level 0>'].astype(str)
f_edu['<ISCED level 1>'] = f_edu['<ISCED level 1>'].astype(str)
f_edu['<ISCED level 2>'] = f_edu['<ISCED level 2>'].astype(str)
f_edu['<ISCED level 3>'] = f_edu['<ISCED level 3>'].astype(str)
```

```python
# Change values from '1's and 'Yes's to the corresponding level to prepare for the melt
f_edu['<ISCED level 0>'] = f_edu['<ISCED level 0>'].replace(regex='1', value='Early childhood')
f_edu['<ISCED level 1>'] = f_edu['<ISCED level 1>'].replace(regex='1', value='Primary')
f_edu['<ISCED level 2>'] = f_edu['<ISCED level 2>'].replace(regex='1', value='Lower secondary')
f_edu['<ISCED level 3>'] = f_edu['<ISCED level 3>'].replace(regex='1', value='Upper secondary')
f_edu['<ISCED level 4>'] = f_edu['<ISCED level 4>'].replace(regex='Yes', value='Post-secondary')
f_edu['<ISCED level 5>'] = f_edu['<ISCED level 5>'].replace(regex='Yes', value='Short-cycle tertiary')
f_edu['<ISCED level 6>'] = f_edu['<ISCED level 6>'].replace(regex='Yes', value='Bachelorâ€™s or equivalent')
```

```python
# Test to see if the columns are all correct before melt
f_edu.sample(10)
```

```python
# Combine all columns into one variable column
f_edu = pd.melt(f_edu, 
                     id_vars=['Country','Student ID', 'Gender',
                            'Mother - Highest Schooling', 
                            'Mother Qualifications - ISCED level 6', 
                            'Mother Qualifications - ISCED level 5A',
                            'Mother Qualifications - ISCED level 5B',
                            'Mother Qualifications - ISCED level 4',
                            'Out-of-School Study Time - Homework',
                            'Out-of-School Study Time - Guided Homework',
                            'Out-of-School Study Time - Personal Tutor',
                            'Out-of-School Study Time - Commercial Company',
                            'Out-of-School Study Time - With Parent',
                            'Learning Time - Mathematics',
                            'Learning Time - Test Language',
                            'Learning Time - Science',
                            'Average_Math_Score',
                            'Average_Reading_Score',
                            'Average_Science_Score',
                            'Average_Total_Score'], var_name='Completed_edu - Father', value_name='Education - Father')
```

```python
# Check to see if the columns from the melt were correctly created
f_edu.sample(10)
```

```python
# Clean all cases from melt where the edu of the father is either '0' or 'No'
f_edu = f_edu[(f_edu['Education - Father'] != '0')] 
f_edu = f_edu[(f_edu['Education - Father'] != 'No')]
```

```python
# Check to see if 0's and No's have been removed
f_edu.sample(70)
```

```python
# Remove 'Father - Highest Schooling' since it is redundant
f_edu = f_edu[(f_edu['Completed_edu - Father'] != 'Father - Highest Schooling')] 
```

```python
# Check to see if correct values remain
f_edu['Education - Father'].value_counts()
```

```python
# Completed_edu is no longer needed since it duplicates the information in 'Education - Father'
f_edu.drop(columns=['Completed_edu - Father'], axis=1, inplace=True)
```

```python
# Check the column has been dropped
f_edu.sample(10)
```

```python
# Check if there are any duplicated values 
f_edu['Student ID'].duplicated().sum()
```

```python
f_edu.dropna(inplace = True)
```

```python
# Change datatypes into category, and put into order from lowest to highest
f_edu['Education - Father'] = f_edu['Education - Father'].astype('category')
f_edu['Education - Father'] = f_edu['Education - Father'].cat.set_categories(['Bachelorâ€™s or equivalent','Short-cycle tertiary','Post-secondary',
                                                                              'Upper secondary',
                                                                              'Lower secondary', 
                                                                              'Primary', 
                                                                              'Early childhood'], ordered=True)
f_edu.sort_values(['Education - Father'], inplace=True)
```

```python
# Remove duplicates and keep last, which is the highest level of education for the student's father
f_edu_clean = f_edu.drop_duplicates(subset=['Student ID'], keep='first')
```

```python
# Ensure there are no duplicates
f_edu_clean['Student ID'].duplicated().sum()
```

```python
# Check the size of the dataframe - 
print(pisa['STIDSTD'].nunique(), f_edu_clean.shape)
```
